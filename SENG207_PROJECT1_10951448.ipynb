# Getting rid of null values:
#### The data shown below show the rows with null values in them.

```python
nan_rows = data[data.isnull().T.any().T]
nan_rows
```

#### The data below shows the rows that are duplicated. 
#### We can see that there are no duplicate rows. 

```python
duplicate_rows = data[data.duplicated()]
duplicate_rows
```

#### we remove the null values and duplicate values by using the dropna() method. Observe that we assign the new dataset with the removed null values as out new dataset. 

```python
data = data.dropna()
```

```python
data
```

#### To prove that the null values have been removed;

```python
nan_rows = data[data.isnull().T.any().T]
nan_rows
```

#### we realize that there is no output and hence the null values have been removed.
#### For the null values, I chose to drop them entirely from the dataset using the dropna() method because I didn't have enough information to input or fill in the missing values. Dropping the null values will also ensure that the data analysis is performed on complete and accurate data.

# Showing Data correlation among columns:

#### This can easily be done using the 'corr()' method. 
##### Below is a representsation and graph of all the correlations between all the columns

```python
corr = data.corr()
corr
```

###### Graph:

```python
fig= plt.figure(figsize= (8,8))
plt.matshow(corr, cmap= 'RdBu', fignum= fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical');
plt.yticks(range(len(corr.columns)), corr.columns);
```

#### The above graph diplays the correlation among all columns of the dataset. the strongest correlations are shown in blue and the weakest in red. We can infer from the graph which columns are strongly related and which ones are not.
